```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
myData = pd.read_csv("Water_Consumption_And_Cost__2013_-_Feb_2022_.csv")
myData.info()
print('-'*100)
myData['Current Charges']=myData['Current Charges'].str.replace(',', '').astype(float)
myData['Consumption (HCF)']=myData['Consumption (HCF)'].str.replace(',', '').astype(float)
myData["Revenue Month"] = pd.to_datetime(myData["Revenue Month"])
myData.info()
print('-'*100)
```

```python
myData.describe()
```

```python
#Full display 
plt.rcParams['figure.figsize'] = (20, 10)
myData.plot(x="Revenue Month", y="Current Charges")
plt.show()
#This is with Consumption. Also, Analyze consumption vs current Charges and other charges 
myData.plot(x="Revenue Month", y="Consumption (HCF)", linestyle = 'dotted')
plt.tight_layout()
plt.show()
```

```python
from matplotlib.dates import MonthLocator, DateFormatter
#Trying to sort the data
plt.rcParams['figure.figsize'] = (20, 15)
myData_sort=myData.sort_values(by='Revenue Month')
ax = myData_sort.plot(x="Revenue Month", y="Current Charges")
#defines the tick location 
ax.xaxis.set_major_locator(MonthLocator())
#defines the label format
ax.xaxis.set_major_formatter(DateFormatter("%m-%y"))
ax.tick_params(axis="x", labelrotation= 90)
plt.tight_layout()
plt.show()
#This is with Consumption. Also, Analyze consumption vs current Charges and other charges 
myData_sort.plot(x="Revenue Month", y="Consumption (HCF)")

plt.show()

```

```python
queens_df= myData_sort.loc[myData['Borough']=='QUEENS']
brooklyn_df= myData_sort.loc[myData['Borough']=='BROOKLYN']
bronx_df= myData_sort.loc[myData['Borough']=='BRONX']
fha_df=myData_sort.loc[myData['Borough']=='FHA']
staten_island_df= myData_sort.loc[myData['Borough']=='STATEN ISLAND']
manhattan_df= myData_sort.loc[myData['Borough']=='MANHATTAN']
```
